## Building the Police BJS Database

```python
## Python packages - you may have to pip install sqlalchemy, sqlalchemy_utils, and psycopg2.
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
import psycopg2
import pandas as pd
```

```python
#In Python: Define a database name (we're using a dataset on births, so I call it 
# birth_db), and your username for your computer (CHANGE IT BELOW). 
dbname = 'police_bjs'
username = 'along528'
pswd = 'password'
## 'engine' is a connection to a database
## Here, we're using postgres, but sqlalchemy can connect to other things too.
engine = create_engine('postgresql://%s:%s@localhost/%s'%(username,pswd,dbname))
print engine.url
# Replace localhost with IP address if accessing a remote server
```

```python
## create a database (if it doesn't exist)
if not database_exists(engine.url):
    create_database(engine.url)
print(database_exists(engine.url))
```

```python
# load a database from CSV
police_csv  = '/Users/along528/Dropbox/insight/project/datasets/police/ICPSR_31161/DS0001/31161-0001-Data.csv'
police_data = pd.DataFrame.from_csv(police_csv)
police_data.columns = map(str.lower, police_data.columns)
police_data[police_data['state']=='NC']
```

```python
police_data.to_sql('police_data_table', engine, if_exists='replace')
```
