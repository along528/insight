## Illinois

```python
df_il = pd.read_sql('SELECT * FROM il_traffic_stops_integrate_year',con)
#use same agency name as survey. Mismatches due mostly to typos in survey
df_il.ix[df_il['agencyname']=='Mount Zion Police','agencyname'] = 'Mt Zion'
df_il.ix[df_il['agencyname']=='Kankakee County Sheriff','agencyname'] = 'Kannakee'
df_il.ix[df_il['agencyname']=='Rosemont Police','agencyname'] = 'Rosemont Public Safety Department'
df_il.ix[df_il['agencyname']=='Berkeley Police','agencyname'] = 'Benkeley'
df_il.ix[df_il['agencyname']=='Winfield Police','agencyname'] = 'Winifield'
df_il.ix[df_il['agencyname']=='Wilmington Police','agencyname'] = 'Willmington'
df_il.ix[df_il['agencyname']=='Wabash County Sheriff','agencyname'] = 'Wabath'
df_il
```

```python
map_survey_to_traffic['IL'] = {}
map_traffic_to_survey['IL'] = {}
nmatch=0
for survey_index,survey_series in survey_df[survey_df['state']=='IL'].iterrows():
    survey_agency = survey_series['agency']
    survey_id     = survey_series['SURVEYID']
    survey_agency = survey_agency.title()
    if "State" in survey_agency: continue #don't consider state patrol
    #if 'Police' in survey_agency
    survey_agency = survey_agency.split('Police').pop(0)
    survey_agency = survey_agency.split('Count').pop(0)
    survey_agency = survey_agency.split(' Co ').pop(0)
    survey_agency = survey_agency.split(' Il').pop(0)
    survey_agency = survey_agency.strip()
    #print survey_agency
    match = False
    for traffic_index, traffic_series in df_il.iterrows():
        traffic_agency = traffic_series['agencyname'].title()
        traffic_agency = traffic_agency.split('Police').pop(0)
        traffic_agency = traffic_agency.split('County').pop(0)
        traffic_agency = traffic_agency.strip()
        traffic_id = traffic_series['agencyid']
        #print "\t",traffic_agency
        if survey_agency==traffic_agency:
            nmatch+=1
            match=True
            break
    if  not match: print survey_agency
    if match: 
            map_survey_to_traffic['IL'][survey_id] = traffic_id
            map_traffic_to_survey['IL'][traffic_id] = survey_id
print "Number of matched agencies =",nmatch
```

Matched all of those in the survey!

```python
pickle.dump(map_survey_to_traffic['IL'],open('map_survey_to_traffic_il.p','wb'))
pickle.dump(map_traffic_to_survey['IL'],open('map_traffic_to_survey_il.p','wb'))
```

## Create combined traffic table with mapped survey id

```python
dbname = 'combined_profiling'
username = 'along528'
pswd = 'password'
## 'engine' is a connection to a database
## Here, we're using postgres, but sqlalchemy can connect to other things too.
engine = create_engine('postgresql://%s:%s@localhost/%s'%(username,pswd,dbname))
print engine.url
```

```python
new_map = {}
for state in map_traffic_to_survey:
     
    for agencyid in map_traffic_to_survey[state]:
        new_map[state+str(agencyid)] = map_traffic_to_survey[state][agencyid]
```

```python
comb_df = pd.read_sql('SELECT * FROM combined_traffic_stops_integrate_year',con)
comb_df
```

```python
comb_df['surveyid'] = comb_df['agencyid'].map(new_map)
comb_df = comb_df.fillna(0)
comb_df = comb_df[comb_df['surveyid']!=0]
comb_df = comb_df.drop('agencyid',axis=1)
comb_df = comb_df.drop('index',axis=1)
comb_df
```

```python
comb_df.to_sql('combined_traffic_stops_integrate_year_surveyid',engine,if_exists='replace')
```
