## Connecticut

```python
df_ct = pd.read_sql('SELECT * FROM ct_traffic_stops_integrate_year',con)
df_ct.ix[df_ct['agencyname']=='Groton Town','agencyname'] = 'Groton'
df_ct.ix[df_ct['agencyname']=='Shelton','agencyname'] = 'Sheldon'
df_ct
```

```python
map_survey_to_traffic['CT'] = {}
map_traffic_to_survey['CT'] = {}
nmatch=0
for survey_index,survey_series in survey_df[survey_df['state']=='CT'].iterrows():
    survey_agency = survey_series['agency']
    survey_id     = survey_series['SURVEYID']
    survey_agency = survey_agency.title()
    if "State" in survey_agency: continue #don't consider state patrol
    survey_agency = survey_agency.split('Police').pop(0)
    survey_agency = survey_agency.split('Connec').pop(0)
    survey_agency = survey_agency.strip()
    #print survey_agency
    match = False
    for traffic_index, traffic_series in df_ct.iterrows():
        traffic_agency = traffic_series['agencyname']
        traffic_id = traffic_series['agencyid']
        #print "\t",traffic_agency
        if survey_agency==traffic_agency:
            nmatch+=1
            match=True
            break
    if  not match: print survey_agency
    if match: 
            map_survey_to_traffic['CT'][survey_id] = traffic_id
            map_traffic_to_survey['CT'][traffic_id] = survey_id
print "Number of matched agencies =",nmatch
```

Matched every CT department in survey!

```python
pickle.dump(map_survey_to_traffic['CT'],open('map_survey_to_traffic_ct.p','wb'))
pickle.dump(map_traffic_to_survey['CT'],open('map_traffic_to_survey_ct.p','wb'))
```
