```python
#counties_surveydf = surveydf[(surveydf['agency'].str.contains('COUNTY')) & (surveydf['agency'].str.contains('SHERIF'))]
counties_surveydf = surveydf[(surveydf['agency'].str.contains('PARISH')) |\
                             (surveydf['agency'].str.contains('COUNTY')) |\
                             (surveydf['agency'].str.contains('SHERIF'))]
counties_surveydf
```

And specifically those relevant for the traffic data

```python
counties_traffic_surveydf = counties_surveydf[(counties_surveydf['state']=='NC') |\
                           (counties_surveydf['state']=='CT') |\
                           (counties_surveydf['state']=='IL')]
```

```python
def get_county_from_agency(agency):
    county = ''
    if ' CO ' in agency:
        county = agency.split(' CO ').pop(0)
    elif ' PARISH ' in agency:
        county = agency.split('PARISH').pop(0)
    else:
        county = agency.split('COUNTY').pop(0)
    if 'SHERIF' in county:
        county = county.split('SHERIF').pop(0)
    county = county.strip()
    return county.title()
zips = []
count=0
for index,series in counties_traffic_surveydf.iterrows():
    name = series['agency']
    county = get_county_from_agency(name)
    city = series['city']
    state = series['state']
    testdf = zipdf[(zipdf['County']==county) & (zipdf['State Abbreviation']==state)]
    #zips+=testdf['Postal Code'].tolist()
    if len(testdf)==0: 
        print "%s,%s,%s,%s" %(name,city,county,state)
        count+=1
    #print survey_name,survey_city,survey_state,len(testdf)
print count
```

## Create Map

```python
testzipdf = pd.DataFrame(zipdf)
testzipdf['surveyid_city'] = -1
testzipdf['surveyid_county'] = -1

for index,zip_series in zipdf.iterrows():
    city = zip_series['Place Name']
    county = zip_series['County']
    state = zip_series['State Abbreviation']
    survey_slice =  surveydf[(surveydf['state']==state) &\
                   ((surveydf['city']==city) |\
                    (surveydf['county']==county))]
    if len(testdf)==0: continue
    #print index,city,county,state
    for _,survey_series in survey_slice.iterrows():
        if survey_series['iscountysheriff']:
            testzipdf.ix[index,'surveyid_county'] = survey_series['SURVEYID']
        else:
            testzipdf.ix[index,'surveyid_city'] = survey_series['SURVEYID']
```
