Goal: Develop a model that predicts whether a biopsied breast cell is benign (not harmful) or malignant (cancerous), given a set of attributes about the cell

```python
## Python packages - you may have to pip install sqlalchemy, sqlalchemy_utils, and psycopg2.

import pandas as pd
# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
pd.set_option('display.max_rows', 20)

import numpy as np
import math
# The usual preamble
%matplotlib inline
%pylab inline
import matplotlib.pyplot as plt
plt.rcParams['axes.color_cycle'] = ['r', 'g', 'b', 'c']
plt.rcParams['lines.color'] = 'r'
plt.rcParams['figure.figsize'] = (15, 5)
from matplotlib.colors import ListedColormap

import seaborn as sns
```

Besides 'id' and 'class', all features are numerical ranging from 1-10. 'class' is a label with 2 meaning benign and 4 meaning malignant

```python
df = pd.read_csv('breast-cancer-wisconsin.data')
df
```

```python
def process(df):
    df = df.set_index('id')
    def label(x):
        if x==2: return False
        elif x==4: return True
        return None
    def replace_question(x):
        if x=='?': return np.nan
        return x
    df['is_malignant'] = df['class'].map(label)
    for column in df.columns.tolist():
        df[column] = df[column].map(replace_question)
    df = df.drop('class',1)
    df = df.dropna()
    for column in df.columns.tolist():
        if 'is_malignant'  in column or 'id' in column: continue
        df[column] = pd.to_numeric(df[column])

    return df
df = process(df)
df
```

For plotting correlations, add a random smearing so we don't overlap all of the points

```python

np.random.seed(20)
def rand(x):
    return x + (np.random.random()-0.5)
def smear(df):
    for column in df.columns.tolist():
        if 'id' in column or 'is_malignant' in column: continue
        df[column] = df[column].map(rand)
    return df
smeared_df = smear(df)
smeared_df
```

Now plot all of the features against eachother and color them by whether or not they labeled as malignant

```python
g = sns.PairGrid(smeared_df,hue='is_malignant')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
g.add_legend();
```

It is clearly possible to get some good separation even just using a few of these correlated variables. For instance, using clump_thickness vs bland_chromatin. Nearly all of the variables are are associated with benign tumors for lower values of the features.

## Modeling

Split the training and testing data

```python
def split_data(data,frac=0.2,random_state=20):
     test = data.sample(frac=frac,random_state=random_state)
     train = data[data.index.isin(test.index.values.tolist())==False]
     return train,test
def get_Xy(df,label_col='is_malignant'):
    return df.drop(label_col,1).as_matrix(),\
           df[label_col].as_matrix()
X,y = get_Xy(df)
train_df,test_df = split_data(df)
train_X,train_y = get_Xy(train_df)
test_X,test_y = get_Xy(test_df)
```

Let's try a simple logistic regression model out of the gate. Build a pipeline for scaling, cross-validation, and training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
scaler = StandardScaler()
classifier = LogisticRegression(penalty='l2')
params = {'C': np.logspace(-3,10,14)}
cv = GridSearchCV(classifier,params,n_jobs=1,cv=10)
log_regr = Pipeline([('scaler',scaler),('cv',cv)])
log_regr.fit(train_X,train_y)
```

```python
log_regr.score(train_X,train_y)
```

```python
log_regr.score(test_X,test_y)
```

This already looks pretty good. Let's check the learning curve to see how things are converging

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and traning learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : integer, cross-validation generator, optional
        If an integer is passed, it is the number of folds (defaults to 3).
        Specific cross-validation objects can be passed, see
        sklearn.cross_validation module for the list of possible objects

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
from sklearn.learning_curve import learning_curve
train_X_scaled = scaler.fit_transform(train_X)
plot_learning_curve(LogisticRegression(), "",train_X_scaled, train_y, cv=20,
                    train_sizes=np.linspace(0.1,1.,5))
```

The train and cross-val scores seem to mostly approach each other, which is a good sign. The precision, recall, and fscore are also all quite good as can be seen below:

```python
from sklearn.metrics import precision_recall_fscore_support
pred_y = log_regr.predict(test_X)

precision,recall,f1,support = precision_recall_fscore_support(test_y,pred_y,average='binary')
print "Precision, tp/(tp+fp) =", precision
print "Recall, tp/(tp+fn) =",recall
print "f1 =",f1
```

## Feature Exploration

It seems like from the earlier study of the features correlations that the model could perform well even using just a couple of features. Let's run the model on just two features at at time and then plot the decision function on each one along with the test score.

```python
plt.rcParams['figure.figsize'] = (25, 25)

cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
columns = train_df.drop('is_malignant',1).columns.tolist()
h=.02 #step size in mesh
for i,col1 in enumerate(columns):
    for j,col2 in enumerate(columns):
        if i>=j: continue
        X_train,y_train = get_Xy(train_df[[col1,col2,'is_malignant']])
        X_test,y_test = get_Xy(test_df[[col1,col2,'is_malignant']])
        x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
        y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
        log_regr_tmp = Pipeline([('scaler',scaler),('cv',cv)])
        log_regr_tmp.fit(X_train,y_train)
        ax = plt.subplot(len(columns), len(columns), (i)*len(columns)+j+1)
        #ax = plt.subplot((i+1)*100+(j+1)*10+1)
        Z = log_regr_tmp.named_steps['cv'].best_estimator_.decision_function(np.c_[xx.ravel(), yy.ravel()])
        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)
        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6)
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xlabel(col1)
        ax.set_ylabel(col2)
        score = log_regr_tmp.score(X_test,y_test)
        ax.set_title('Test Score=%3.2f' % (score))
        ax.set_xticks(())
        ax.set_yticks(())

```

Indeed, there are several pairs of variables which perform almost as well as the model with all variables included. Note that the decision function is very simple. The model accuracy could be improved even further by including a non-linear model like an SVM or tree-based algorithm, for example. But I've already got a lot of power out of the model.
