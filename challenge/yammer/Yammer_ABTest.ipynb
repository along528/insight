```python
import math
# The usual preamble
%matplotlib inline
%pylab inline
import matplotlib.pyplot as plt
plt.rcParams['axes.color_cycle'] = ['r', 'g', 'b', 'c']
plt.rcParams['lines.color'] = 'r'
plt.rcParams['figure.figsize'] = (15, 5)
import datetime
```

```python
name = 'merged_all'
days_test = df[name][df[name]['experiment_group']=='test_group']['activated_at_datetime'].map(lambda x: (x - np.datetime64('2014-06-01')).days).values.tolist()
days_control = df[name][df[name]['experiment_group']=='control_group']['activated_at_datetime'].map(lambda x: (x - np.datetime64('2014-06-01')).days).values.tolist()
ranges=[-100,30]
plt.hist(days_control,bins=30,range=ranges) #,days_control)
plt.hist(days_test,bins=30,alpha=.5,range=ranges) #,days_control)
plt.xlim(ranges)
plt.ylim([0,200])
plt.legend(['Control','Test'])
plt.xlabel('Days since start of test')
plt.ylabel('Number of new users')
```

compute p-values

```python
from scipy.stats import ttest_ind
```

```python
name = 'merged_corrected'
test = df[name][df[name]['experiment_group']=='test_group']['count'].values.tolist()
control = df[name][df[name]['experiment_group']=='control_group']['count'].values.tolist()
ttest_ind(test,control,equal_var=False).pvalue
plt.hist(test,bins=60) #,bins=0)
plt.hist(control,bins=60,alpha=.4) #,bins=80,alpha=.4)
#plt.xlim([0,100])
np.mean(control),np.mean(test)
```
